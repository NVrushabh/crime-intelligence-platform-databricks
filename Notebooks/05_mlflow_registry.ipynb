### 📝 Register Models into MLflow Model Registry
Registering our 3 best finished models with official names and versions.

```python
import mlflow

mlflow.set_registry_uri("databricks-uc")

runs_to_register = {
    "workspace.default.arrest_likelihood_scorer": (
        "317da5b60be841e2b220557c365dc0df",
        "arrest_likelihood_model"
    ),
    "workspace.default.crime_type_classifier": (
        "a912d51c3c504f2d8b722168151dd3bd",
        "crime_type_classifier"
    ),
    "workspace.default.crime_hotspot_predictor": (
        "73745bf9cd4d47f08ff67f15e65849f3",
        "crime_hotspot_predictor"
    )
}

for model_name, (run_id, artifact_path) in runs_to_register.items():
    model_uri = f"runs:/{run_id}/{artifact_path}"

    try:
        registered = mlflow.register_model(
            model_uri=model_uri,
            name=model_name,
            await_registration_for=300
        )

        print(f"✅ Registered: {model_name}")
        print(f"   Version: {registered.version}")
        print(f"   Status: {registered.status}")
        print()

    except Exception as e:
        print(f"⚠️ Could not register {model_name}")
        print(str(e)[:500])
        print()
```
