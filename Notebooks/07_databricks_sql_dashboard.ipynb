## 📊 Databricks SQL Dashboard - Crime Intelligence Visualization
This notebook prepares dashboard-ready analytics views and SQL queries from Gold layer tables to visualize crime patterns, hotspot districts, arrest likelihood, and temporal crime trends.

### 🥇 Top Crime Types Analysis
Creating dashboard-ready analytics for the most common crime categories and their arrest rates across the city.

```python
top_crimes_dashboard = spark.sql("""

SELECT
    Primary_Type,
    COUNT(*) AS total_crimes,
    ROUND(AVG(CASE WHEN Arrest = true THEN 1 ELSE 0 END) * 100, 2) AS arrest_rate_pct

FROM workspace.default.silver_crimes

GROUP BY Primary_Type
ORDER BY total_crimes DESC
LIMIT 15

""")

display(top_crimes_dashboard)
```

### ⏰ Crime Activity by Hour
Analyzing hourly crime distribution to identify peak crime activity periods across the day.

```python
crime_by_hour_dashboard = spark.sql("""

SELECT
    Hour,
    COUNT(*) AS total_crimes,
    
    ROUND(
        AVG(CASE WHEN Arrest = true THEN 1 ELSE 0 END) * 100,
        2
    ) AS arrest_rate_pct

FROM workspace.default.silver_crimes

GROUP BY Hour
ORDER BY Hour

""")

display(crime_by_hour_dashboard)
```

### 📍 District Crime Hotspots
Identifying districts with the highest crime concentration and comparing arrest activity across districts.

```python
district_hotspot_dashboard = spark.sql("""

SELECT
    District,
    COUNT(*) AS total_crimes,

    ROUND(
        AVG(CASE WHEN Arrest = true THEN 1 ELSE 0 END) * 100,
        2
    ) AS arrest_rate_pct

FROM workspace.default.silver_crimes

WHERE District IS NOT NULL

GROUP BY District
ORDER BY total_crimes DESC

""")

display(district_hotspot_dashboard)
```

### 🌙 Weekend vs Night Crime Patterns
Comparing crime activity across weekends and nighttime periods to identify behavioral crime trends.

```python
weekend_night_dashboard = spark.sql("""

SELECT
    Is_Weekend,
    Is_Night,

    COUNT(*) AS total_crimes,

    ROUND(
        AVG(CASE WHEN Arrest = true THEN 1 ELSE 0 END) * 100,
        2
    ) AS arrest_rate_pct

FROM workspace.default.silver_crimes

GROUP BY Is_Weekend, Is_Night

ORDER BY total_crimes DESC

""")

display(weekend_night_dashboard)
```

### 📈 Seasonal Crime Trends
Analyzing how crime activity changes across different seasons throughout the year.

```python
seasonal_crime_dashboard = spark.sql("""

SELECT
    Season,

    COUNT(*) AS total_crimes,

    ROUND(
        AVG(CASE WHEN Arrest = true THEN 1 ELSE 0 END) * 100,
        2
    ) AS arrest_rate_pct

FROM workspace.default.silver_crimes

GROUP BY Season

ORDER BY total_crimes DESC

""")

display(seasonal_crime_dashboard)
```

### 🚔 Arrest Likelihood by Crime Type
Comparing arrest percentages across major crime categories to identify enforcement and resolution patterns.

```python
arrest_likelihood_dashboard = spark.sql("""

SELECT
    Primary_Type,

    COUNT(*) AS total_crimes,

    ROUND(
        AVG(CASE WHEN Arrest = true THEN 1 ELSE 0 END) * 100,
        2
    ) AS arrest_rate_pct

FROM workspace.default.silver_crimes

GROUP BY Primary_Type

HAVING total_crimes > 1000

ORDER BY arrest_rate_pct DESC

""")

display(arrest_likelihood_dashboard)
```

### 🕸️ Crime Series & Pattern Connections
Visualizing connected crime patterns and repeated incident relationships identified through graph analytics.

```python
crime_series_dashboard = spark.sql("""

SELECT
    block,
    crime_type,
    connections

FROM workspace.default.gold_crime_series_patterns

ORDER BY connections DESC

""")

display(crime_series_dashboard)
```
